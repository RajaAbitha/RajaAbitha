Loading the Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/train_loan.csv')
df
```

```python
print(df.isnull().sum())
```

Missing value Handling

```python
df['Gender']=df['Gender'].fillna(df['Gender'].mode()[0])
df['Married']=df['Married'].fillna(df['Married'].mode()[0])
df['Dependents']=df['Dependents'].fillna(df['Dependents'].mode()[0])
df['Self_Employed']=df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
df['LoanAmount']=df['LoanAmount'].fillna(df['LoanAmount'].mean())
df['Loan_Amount_Term']=df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].median())
df['Credit_History']=df['Credit_History'].fillna(df['Credit_History'].mode()[0])
print(df.isnull().sum())
#print(df)
```

#Outlier Detection

```python
df.describe()
```

```python
upper_limit = df['LoanAmount'].quantile(0.95)  # 95th percentile
lower_limit = df['LoanAmount'].quantile(0.05)  # 5th percentile

df['LoanAmount'] = np.where(df['LoanAmount'] > upper_limit, upper_limit,
                                   np.where(df['LoanAmount'] < lower_limit, lower_limit,
                                            df['LoanAmount']))
df.describe()
```

```python
upper_limit = df['Loan_Amount_Term'].quantile(0.80)
lower_limit = df['Loan_Amount_Term'].quantile(0.20)
df['Loan_Amount_Term'] = np.where(df['Loan_Amount_Term'] > upper_limit, upper_limit,
                                   np.where(df['Loan_Amount_Term'] < lower_limit, lower_limit,
                                            df['Loan_Amount_Term']))
df.describe()
```

#Scaling

```python
standard_scaler=StandardScaler()
numerical_features = df.select_dtypes(include=np.number).columns
df_numerical = df[numerical_features]
df_standardized = pd.DataFrame(standard_scaler.fit_transform(df_numerical),
                                 columns=df_numerical.columns,
                                 index=df.index)
df_final=pd.concat([df_standardized,df.drop(columns=numerical_features)],axis=1)
print(df_final)
```
