```python
# import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('/content/diabetes(1).csv')
```

```python
df.head()
```

```python
# Checking missing values
df.isnull().sum()
```

```python
# Replace Zeros with NaN

cols_with_zero_as_nan = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
df[cols_with_zero_as_nan] = df[cols_with_zero_as_nan].replace(0, np.nan)
```

```python
df.head()
```

```python
# Fill missing values with median
df[cols_with_zero_as_nan] = df[cols_with_zero_as_nan].fillna(df[cols_with_zero_as_nan].median())

```

```python
df.head()
```

```python
x=df.drop('Outcome',axis=1)
y=df['Outcome']
```

```python
# split the dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
print(y_pred)
```

```python
# Evaluate the model
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,classification_report
print(accuracy_score(y_test,y_pred))
```

```python
# confusion metrix
print(confusion_matrix(y_test,y_pred))
```

```python
# classification report
print(classification_report(y_test,y_pred))
```

```python
# Random forest model
from sklearn.ensemble import RandomForestClassifier
model1=RandomForestClassifier()
model1.fit(x_train,y_train)
```

```python
# Evaluation
y_pred1=model1.predict(x_test)
print(accuracy_score(y_test,y_pred1))
```

```python
# confusion metrix
print(confusion_matrix(y_test,y_pred1))
```

```python
# Classification report
print(classification_report(y_test,y_pred1))
```
