```python
# 1 Read the dataset
import pandas as pd
data=pd.read_csv('/content/diabetes.csv')

```

```python
# 2 Display columns in the dataset
print(data.columns)
print(data.head())
```

```python
# Display mean of each column in the dataset
print(data.mean())
```

```python
# Check null values
print(data.isnull().sum())
```

```python
# perform meaningful visualization
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(data.corr())
plt.show()
```

```python
sns.boxplot(data['BMI'])
plt.show()
```

```python
# Draw scatter plot
sns.scatterplot(x=data['BMI'],y=data['Age'])
plt.show()
```

```python
plt.bar(data['Age'],data['BMI'])
plt.show()
```

```python
# BAR CHART vs Age and blood pressure
plt.bar(data['Age'],data['BloodPressure'])
plt.show()
```

```python
plt.bar(data['Age'],data['Pregnancies'], color='red')
plt.show()
```

```python
plt.stem(data['Age'],data['Pregnancies'])
plt.show()
```
