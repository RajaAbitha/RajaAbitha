```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read the dataset

train_data=pd.read_csv('/content/train_data.csv')
test_data=pd.read_csv('/content/test_data.csv')
```

```python
# find the null values
train_data.isnull().sum()
```

```python
# save employee id before it drop
train_id=train_data['employee_id']
test_id=test_data['employee_id']
```

```python
# Drop employee id column

train_data.drop('employee_id',axis=1,inplace=True)
```

```python
# drop employee id from test data
test_data.drop('employee_id',axis=1,inplace=True)
```

```python
train_data['is_promoted'].unique()
```

```python
# Handle categorical column with label encoding
from sklearn.preprocessing import LabelEncoder
cat_cols=['department','region','education','gender','recruitment_channel']
le_dict={}
for col in cat_cols:
  le=LabelEncoder()
  train_data[col]=le.fit_transform(train_data[col])
  test_data[col]=le.transform(test_data[col])
  le_dict[col]=le
```

```python
# handle missing values
train_data.isnull().sum()
```

```python
# fill the previous year rating with median value

mean_value=train_data['previous_year_rating'].mean()
train_data['previous_year_rating'] = train_data['previous_year_rating'].fillna(mean_value)
test_data['previous_year_rating'] = test_data['previous_year_rating'].fillna(mean_value)
```

```python
train_data.isna().sum()
```

```python
X=train_data.drop('is_promoted',axis=1)
y=train_data['is_promoted']
```

```python
test_data.isna().sum()
```

```python
# scale the feature
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(X)
test_data=sc.transform(test_data)
```

```python
# train the model

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

log_model=LogisticRegression()
log_model.fit(X,y)
```

```python
# prediction

print('train accuracy',accuracy_score(y,log_model.predict(X)))
```
